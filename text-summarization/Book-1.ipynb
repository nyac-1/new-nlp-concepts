```python
import numpy as np
import pandas as pd
```

```python
columns = ['headline', 'text', 'class']
df = pd.DataFrame(columns=columns)
```

```python
import os

directory = './bbc'
 
for filename in os.scandir(directory):
    if filename.is_dir():
        text_class = filename.path.split("/")[-1]
        for file in os.scandir(filename.path):
            if file.is_file():
                with open(file.path) as f:
                    lines = f.readlines()
                    headline = lines[0]
                    lines.remove(lines[0])
                    try:
                        lines.remove("\n")
                    except:
                        pass
                    lines = "".join(lines)
                    
                    temp = {
                        "headlineline, lines, text_class]
                    
                    df = df.append(temp, ignore_index=True)
            break
    
```
